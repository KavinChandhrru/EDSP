```python
!pip install librosa matplotlib
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np

# Specify the file paths to your audio files
audioFilePath1 = '/content/Broken 2.7_trimmed_center.wav'
audioFilePath2 = '/content/Healthy 3_trimmed_center.wav'  # Replace with the path to your second audio file

# Read the first audio file
y1, sr1 = librosa.load(audioFilePath1, sr=None)

# Read the second audio file
y2, sr2 = librosa.load(audioFilePath2, sr=None)

# Calculate the time vectors
t1 = np.arange(0, len(y1)) / sr1
t2 = np.arange(0, len(y2)) / sr2

# Plot the Amplitude vs. Time graph for the first audio file
plt.figure()
plt.plot(t1, y1, label='Broken - 2.7', color='black')

# Plot the Amplitude vs. Time graph for the second audio file
plt.plot(t2, y2, label='Healthy - 3', color='orange')

plt.xlabel('Time (seconds)')
plt.ylabel('Amplitude')
plt.title('Amplitude vs. Time Comparison')
plt.legend()  # Show legend to distinguish between the two audio files
plt.grid(True)

# Show the plot
plt.show()
```

# Amplitude VS Frequency

Broken(2.7)A & Healthy(3)A rotor

```python
!pip install librosa matplotlib
import librosa
import matplotlib.pyplot as plt
import numpy as np

# Specify the file paths to your audio files
audioFilePath1 = '/content/Broken 2.7_trimmed_center.wav'
audioFilePath2 = '/content/Healthy 3_trimmed_center.wav'  # Replace with the path to your second audio file

# Read the first audio file
y1, sr1 = librosa.load(audioFilePath1, sr=None)

# Read the second audio file
y2, sr2 = librosa.load(audioFilePath2, sr=None)

# Calculate the Fast Fourier Transform (FFT) for the first audio file
fft_result1 = np.fft.fft(y1)
magnitude1 = np.abs(fft_result1)
frequency1 = np.fft.fftfreq(len(magnitude1), d=1/sr1)

# Keep only the positive frequencies
positive_frequencies1 = frequency1[:len(frequency1)//2]
positive_magnitude1 = magnitude1[:len(magnitude1)//2]

# Calculate the Fast Fourier Transform (FFT) for the second audio file
fft_result2 = np.fft.fft(y2)
magnitude2 = np.abs(fft_result2)
frequency2 = np.fft.fftfreq(len(magnitude2), d=1/sr2)

# Keep only the positive frequencies
positive_frequencies2 = frequency2[:len(frequency2)//2]
positive_magnitude2 = magnitude2[:len(magnitude2)//2]

# Plot the Frequency vs. Amplitude graph for both audio files
plt.figure(figsize=(12, 6))
plt.plot(positive_frequencies1, positive_magnitude1, label='Broken - 2.7', color='black')
plt.plot(positive_frequencies2, positive_magnitude2, label='Healthy - 3', color='orange')

plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude')
plt.title('Frequency vs. Amplitude Comparison')
plt.legend()  # Show legend to distinguish between the two audio files
plt.grid(True)
plt.show()
```

# Kurtosis analysis - Time domain

Broken rotor - 2.7

```python
import librosa
from scipy.stats import kurtosis
import matplotlib.pyplot as plt
import seaborn as sns

# Function to calculate kurtosis of an audio signal
def calculate_kurtosis(y, sr):
    # Calculate kurtosis
    kurt = kurtosis(y)

    return kurt

# Function to interpret kurtosis
def interpret_kurtosis(kurtosis_value):
    if kurtosis_value > 3:
        return "Leptokurtic Distribution"
    elif kurtosis_value < 3:
        return "Platykurtic Distribution"
    else:
        return "Mesokurtic Distribution"

# Specify the path to your audio file
audio_file_path = '/content/Broken 2.7_trimmed_center.wav'  # Replace with the path to your audio file

# Load the audio signal
y, sr = librosa.load(audio_file_path, sr=None)

# Calculate kurtosis
kurtosis_value = calculate_kurtosis(y, sr)

# Interpret kurtosis
kurtosis_interpretation = interpret_kurtosis(kurtosis_value)

# Display the result
print(f"Kurtosis value: {kurtosis_value}")
print(f"Distribution type: {kurtosis_interpretation}")

# Plot the audio signal
plt.figure(figsize=(12, 4))
librosa.display.waveshow(librosa.effects.preemphasis(y), sr=sr)
plt.title('Audio Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.show()

# Plot the histogram
plt.figure(figsize=(12, 4))
sns.histplot(y, bins=50, kde=True)
plt.title('Histogram of Audio Signal')
plt.xlabel('Amplitude')
plt.ylabel('Frequency')
plt.show()
```

# Skewness analysis

Broken rotor - 2.7


```python
import librosa
from scipy.stats import skew
import matplotlib.pyplot as plt
import librosa.display
import numpy as np

# Function to perform skewness analysis
def skewness_analysis(audio_file):
    # Load the audio signal
    y, sr = librosa.load(audio_file, sr=None)

    # Calculate skewness
    skewness_values = []
    for i in range(0, len(y), 1024):  # Adjust the window size as needed
        segment = y[i:i+1024]
        skewness_values.append(skew(segment))

    # Determine skewness direction
    overall_skewness = skew(y)
    skewness_direction = 'Positive Skewness' if overall_skewness > 0 else 'Negative Skewness' if overall_skewness < 0 else 'Zero Skewness'

    return y, sr, skewness_values, overall_skewness, skewness_direction

# Specify the path to your audio file
audio_file_path = '/content/Broken 2.7_trimmed_center.wav'  # Replace with the path to your audio file

# Perform skewness analysis
y, sr, skewness_values, overall_skewness, skewness_direction = skewness_analysis(audio_file_path)

# Display the result
print(f"Overall Skewness value: {overall_skewness}")
print(f"Overall Skewness direction: {skewness_direction}")

# Plot the skewness curve
time = np.arange(0, len(y)/sr, 1024/sr)  # Adjust the window size as needed
plt.figure(figsize=(12, 4))
plt.plot(time[:len(skewness_values)], skewness_values)
plt.title('Skewness Curve')
plt.xlabel('Time (s)')
plt.ylabel('Skewness')
plt.show()

# Plot the original audio signal
plt.figure(figsize=(12, 4))
librosa.display.waveshow(librosa.effects.preemphasis(y), sr=sr)
plt.title('Original Audio Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.show()
```
