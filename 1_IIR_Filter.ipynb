```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np
from scipy import signal
```

## Audio Signal -Input

```python
# Specify the file names and corresponding labels
file_names = ['/content/2.7 broken table.wav', '/content/3 Amps table.wav', '/content/3.3 broken motor.wav', '/content/3.3 Amps table.wav']
labels = ['2.7 A broken', '3 A Healthy', '3.3 A Broken', '3.3 A Healthy']

# Set up the plot
plt.figure(figsize=(12, 10))

for i, file_name in enumerate(file_names):
    # Load the audio
    y, sr = librosa.load(file_name, sr=None)

    # Plot the Time-Domain Waveform
    plt.subplot(4, 1, i + 1)
    librosa.display.waveshow(y, sr=sr)
    plt.title(f'Input Signal - {labels[i]}')
    plt.xlabel('Time (s)')
    plt.ylabel('Amplitude')

# Add space between subplots
plt.subplots_adjust(hspace=0.5)

plt.show()
```

## Noise Reduction using IIR Filter

```python
#IIR Filter
import librosa
import librosa.display
import numpy as np
import matplotlib.pyplot as plt



file_names = ['Broken 2.7_trimmed_center.wav']
labels = ['Broken Rotor (2.7 amps)']

for i, file_name in enumerate(file_names):
    # Analyze the audio
    y, sr = librosa.load(file_name, sr=None)

    # Design a low-pass IIR filter
    cutoff_freq = 700  # Cutoff frequency in Hertz
    filter_order = 4  # Filter order

    # Normalize the frequency
    Wn = cutoff_freq / (0.5 * sr)

    # Design the low-pass Butterworth filter
    b, a = signal.butter(filter_order, Wn, 'low')

    # Apply the IIR filter to the audio signal in the time domain
    filtered_audio_time = signal.lfilter(b, a, y)

    # Time Domain Representation - Unfiltered
    plt.figure(figsize=(10, 6))
    plt.subplot(2, 1, 1)
    librosa.display.waveshow(y, sr=sr)
    plt.title(f'Time Domain - Unfiltered {labels[i]}')
    plt.xlabel('Time (seconds)')
    plt.ylabel('Amplitude')

    # Time Domain Representation - Filtered
    plt.subplot(2, 1, 2)
    librosa.display.waveshow(filtered_audio_time, sr=sr)
    plt.title(f'Filtered Time Domain - {labels[i]}')
    plt.xlabel('Time (seconds)')
    plt.ylabel('Amplitude')

    # Add space between subplots
    plt.subplots_adjust(hspace=0.5)

    plt.show()

```

## Peak Amplitude Analysis

```python
import librosa
import librosa.display
import numpy as np
import matplotlib.pyplot as plt
# Function to design a low-pass Butterworth filter
def butter_lowpass_filter(data, cutoff_freq, sample_rate, filter_order=4):
    nyquist = 0.5 * sample_rate
    normal_cutoff = cutoff_freq / nyquist
    b, a = butter(filter_order, normal_cutoff, btype='low', analog=False)
    y_filtered = lfilter(b, a, data)
    return y_filtered

# Specify the file names and corresponding labels
file_names = ['2.7 broken table.wav', '3 Amps table.wav']
labels = ['Broken Rotor (2.7 amps)', 'Healthy Motor (3 amps)']

for i, file_name in enumerate(file_names):
    # Load the audio
    y, sr = librosa.load(file_name, sr=None)

    # Design a low-pass IIR filter
    cutoff_freq = 1000  # Cutoff frequency in Hertz
    filtered_audio = butter_lowpass_filter(y, cutoff_freq, sr)

    # Analyze the original audio and find the peak amplitude
    peak_amplitude_original = np.max(np.abs(y))

    # Analyze the filtered audio and find the peak amplitude
    peak_amplitude_filtered = np.max(np.abs(filtered_audio))

    # Plot the Time Domain Representation - Unfiltered
    plt.figure()
    plt.subplot(2, 1, 1)
    librosa.display.waveshow(y, sr=sr)
    plt.title(f'Time Domain - Unfiltered {labels[i]}')
    plt.xlabel('Time (seconds)')
    plt.ylabel('Amplitude')

    # Plot the Time Domain Representation - Filtered
    plt.subplot(2, 1, 2)
    librosa.display.waveshow(filtered_audio, sr=sr)
    plt.title(f'Time Domain - Filtered {labels[i]}')
    plt.xlabel('Time (seconds)')
    plt.ylabel('Amplitude')

    # Display the peak amplitudes
    print(f'Peak Amplitude for {labels[i]} (Unfiltered): {peak_amplitude_original}')
    print(f'Peak Amplitude for {labels[i]} (Filtered): {peak_amplitude_filtered}')
    plt.subplots_adjust(hspace=0.5)
plt.show()
```

## FFT - Frequency Domain Analysis

```python
# FFT Analysis

import librosa
import librosa.display
import numpy as np
import matplotlib.pyplot as plt

# Specify the file names and corresponding labels
file_names = ['Healthy 3_trimmed_center.wav']
labels = ['Healthy Rotor (3 amps)']

for i, file_name in enumerate(file_names):
    # Load the audio
    y, sr = librosa.load(file_name, sr=None)

    # Design a low-pass IIR filter
    cutoff_freq = 700  # Cutoff frequency in Hertz
    filter_order = 4  # Filter order

    # Normalize the frequency
    Wn = cutoff_freq / (0.5 * sr)

    # Design the low-pass Butterworth filter
    b, a = signal.butter(filter_order, Wn, 'low')

    # Apply the IIR filter to the audio signal in the time domain
    filtered_audio_time = signal.lfilter(b, a, y)

    # Compute the frequency domain representations
    Y = np.fft.fft(y)
    filtered_audio_freq = signal.lfilter(b, a, y)
    Y_filtered = np.fft.fft(filtered_audio_freq)

    # Plot the Frequency Domain Representation - Unfiltered
    plt.figure(figsize=(10, 8))
    plt.subplot(2, 1, 1)
    f = np.fft.fftfreq(len(Y), 1/sr)
    plt.plot(f, np.abs(Y))
    plt.title(f'Frequency Domain - Unfiltered {labels[i]}')
    plt.xlabel('Frequency (Hz)')
    plt.ylabel('Amplitude')
    plt.xlim([0, 2000])

    # Plot the Frequency Domain Representation - Filtered
    plt.subplot(2, 1, 2)
    f_filtered = np.fft.fftfreq(len(Y_filtered), 1/sr)
    plt.plot(f_filtered, np.abs(Y_filtered))
    plt.title(f'Filtered Frequency Domain - {labels[i]}')
    plt.xlabel('Frequency (Hz)')
    plt.ylabel('Amplitude')
    plt.xlim([0, 2000])

    # Add space between subplots
    plt.subplots_adjust(hspace=0.5)

    plt.show()


```

## FFT Analysis Comparison

```python
import librosa
import librosa.display
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import wavfile
import scipy.signal as signal

# Specify the file names and corresponding labels
file_names = ['Broken 2.7_trimmed_center.wav', 'Healthy 3_trimmed_center.wav']
labels = ['Broken Rotor (2.7 amps)', 'Healthy Motor (3 amps)']

# Create a figure for the plot
plt.figure(figsize=(10, 8))

for i, file_name in enumerate(file_names):
    # Load the audio
    y, sr = librosa.load(file_name, sr=None)

    # Design a low-pass IIR filter
    cutoff_freq = 700  # Cutoff frequency in Hertz
    filter_order = 4  # Filter order

    # Normalize the frequency
    Wn = cutoff_freq / (0.5 * sr)

    # Design the low-pass Butterworth filter
    b, a = signal.butter(filter_order, Wn, 'low')

    # Apply the IIR filter to the audio signal in the time domain
    filtered_audio_time = signal.lfilter(b, a, y)

    # Compute the frequency domain representations
    Y = np.fft.fft(y)
    filtered_audio_freq = signal.lfilter(b, a, y)
    Y_filtered = np.fft.fft(filtered_audio_freq)

    # Plot the Frequency Domain Representation - Unfiltered
    plt.subplot(2, 1, 1)
    f = np.fft.fftfreq(len(Y), 1/sr)
    plt.plot(f, np.abs(Y), label=f'Unfiltered - {labels[i]}', alpha=0.7)

    # Plot the Frequency Domain Representation - Filtered
    plt.subplot(2, 1, 2)
    f_filtered = np.fft.fftfreq(len(Y_filtered), 1/sr)
    plt.plot(f_filtered, np.abs(Y_filtered), label=f'Filtered - {labels[i]}', alpha=0.7)

# Add legends, titles, and labels
plt.subplot(2, 1, 1)
plt.title('Frequency Domain - Unfiltered')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude')
plt.xlim([0, 2000])
plt.legend()

plt.subplot(2, 1, 2)
plt.title('Filtered Frequency Domain')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude')
plt.xlim([0, 2000])
plt.legend()

# Add space between subplots
plt.subplots_adjust(hspace=0.5)

plt.show()
```
